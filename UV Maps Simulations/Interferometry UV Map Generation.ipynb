```python
import numpy as np
import astropy as astropy
from astropy.coordinates import SkyCoord, EarthLocation, AltAz
from astropy.time import Time
import astropy.units as unit
import matplotlib.pyplot as plt
from astropy.coordinates import get_sun
```

# Earth Rotation Synthesis: UV Coverage Simulation for the Sun


The code below simulates the rotation of the Earth as the Sun moves across the sky. The observation location is chosen near the South African Astronomical Observatory (SAAO) in Observatory, Cape Town, with coordinates approximately $33^\circ 56' 02.5'' \, \text{S}, \, 18^\circ 28' 35.7'' \, \text{E}$ at an altitude of 1000 meters.

Starting from the current time, the source (the Sun) moves through a 12-hour cycle, generating the first half of the UV coverage track. By the Hermitian symmetry property of the visibility function, every measurement at $(u, v)$ has a mirrored counterpart at $(-u, -v)$ with a phase reversal.

Due to the Sun’s intense and broadband emission, its power spectral density can be modeled as Gaussian noise with an evenly spread broadband spectrum. However, for this application, the neutral hydrogen ($\mathrm{H\,I}$) line frequency at 1420 MHz is used as the central frequency, with an ultimate goal of a 50 MHz bandwidth to capture Doppler shifts and spectral patterns around this frequency.


```python
location = EarthLocation(
    lat=-(33 + 55/60 + 58.2312/3600) * unit.deg,
    lon=(18 + 28/60 + 35.817/3600) * unit.deg,
    height=25 * unit.m
)

times = Time.now()+ np.linspace(0, 12, 100)*unit.hour  # 12-hour track
#start_time = Time("2025-08-09 17:49:00")  

# Generate 12-hour track with 100 samples
#times = start_time + np.linspace(0, 12, 100) * unit.hour
u_coords, v_coords = [], []
sun = get_sun(times)
source = sun
wavelength = 3e8/1420e6  # For HI line
```

# Baseline and Coordinate Transformation for a Two-Antenna System 

Consider a two-antenna interferometer system with a single geometric baseline. As the Earth rotates relative to the celestial source, this baseline projects into multiple effective baselines. To accurately characterize the antenna spacing along the East-West direction, a coordinate transformation is required.

This transformation maps the local right-handed horizon coordinate system — East, North, and Up $E, N, U$ — into a global coordinate system $x, y, z$ aligned with the Earth's rotational axis. The local horizon coordinates depend on the antenna’s position on the Earth’s surface, whereas the global coordinate system remains fixed relative to the Earth’s rotation axis, thus providing a more consistent frame of reference for analysis.

The transformation from the local ENU system to the global XYZ system is expressed as:

$$
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
0 & -\sin \lambda & \cos \lambda \\
1 & 0 & 0 \\
0 & \cos \lambda & \sin \lambda
\end{bmatrix}
\begin{bmatrix}
E \\
N \\
U
\end{bmatrix}
$$

where

- $E, N, U$ denote the local East-North-Up coordinates,  
- $x, y, z$ denote the global XYZ coordinates fixed relative to the Earth’s rotation,  
- $\lambda$ represents the observer’s latitude.

For the interferometer system under consideration, the geometric baseline lies purely along the East-West direction. Consequently, the \(x\) and \(z\) components of the baseline vector are zero, and the \(y\)-component corresponds directly to the baseline length.

Accordingly, the baseline vector can be represented as:

$$
\mathbf{B} = 
\begin{bmatrix}
B_x \\
B_y \\
B_z
\end{bmatrix}
=
\begin{bmatrix}
0 \\
20 \\
0
\end{bmatrix} \quad \text{(meters)}
$$

This represents a **20-meter pure East-West baseline**.

```python
baseline_vector = np.array([ 20,0,0.0])  # 20 m E-W baseline 
```

# Baseline Projection: Hour Angle, Declination, and UVW Coordinate
As the Earth rotates, the apparent position of a celestial source changes relative to the observer’s location. To accurately model the apparent position of the source which, for this simulation, is the Sun  two key parameters are required: the **hour angle** and the **declination**.

The **hour angle** $h_0$ is defined as the angular distance between the observer’s local meridian and the position of the source, measured in degrees (Boujoudar et al., 2024). It represents the time elapsed since the source last crossed the observer’s meridian, expressed as an angular measure. Formally, it is given by:

$$
h_0 = \mathrm{LST} - \alpha
$$

where

- $\mathrm{LST}$ is the local sidereal time at the observer’s location,  
- $\alpha$ is the right ascension of the source.

The **declination** $\delta$ is the angular distance of the source north or south of the celestial equator.

These parameters are used to compute the $u, v, w$ coordinates, which describe the projection of the interferometer baseline relative to the source direction.

The transformation from Earth-fixed baseline coordinates $(B_x, B_y, B_z)$ to celestial $(u, v, w)$ coordinates is given by:

$$
\begin{bmatrix}
u \\
v \\
w
\end{bmatrix}
= \frac{1}{\lambda}
\begin{bmatrix}
\sin h_0 & \cos h_0 & 0 \\
-\sin \delta \cos h_0 & \sin \delta \sin h_0 & \cos \delta \\
\cos \delta \cos h_0 & -\cos \delta \sin h_0 & \sin \delta
\end{bmatrix}
\begin{bmatrix}
B_x \\
B_y \\
B_z
\end{bmatrix}
$$

where:

- $\lambda$ is the observation wavelength (21 cm Hydrogen line),  
- $h_0$ is the hour angle,  
- $\delta$ is the declination of the phase center.





```python

for t in times:
    # Get source position and hour angle
    source_at_t = get_sun(t)
    lst = t.sidereal_time('mean', longitude=location.lon)
    ha = np.deg2rad((t.sidereal_time('mean', longitude=location.lon) - source_at_t.ra))
    dec = np.deg2rad(source_at_t .dec)
    
    # UVW projection (Thompson et al. "Interferometry and Synthesis in Radio Astronomy")
    u = (baseline_vector[0]*np.sin(ha) + baseline_vector[1]*np.cos(ha))
    v = (-baseline_vector[0]*np.sin(dec)*np.cos(ha) + 
          baseline_vector[1]*np.sin(dec)*np.sin(ha) + 
          baseline_vector[2]*np.cos(dec))
    
    u_coords.append(u)
    v_coords.append(v)

# Convert to wavelengths
u_coords = np.array(u_coords)/wavelength
v_coords = np.array(v_coords)/wavelength
```

# UV Coverage Plot


```python
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(6, 6), dpi=300)  

# Scatter plot
plt.scatter(u_coords, v_coords, s=5, color='blue', alpha=0.7, label='UV points')

# mirrored points for complete coverage
plt.scatter(-u_coords, -v_coords, s=5, color='red', alpha=0.7, label='Conjugate UV points')

print(np.max(u_coords))
print(np.max(v_coords))
plt.title('Simulated UV Coverage from Earth Rotation Relative to the Sun', fontsize=14, fontweight='bold')
plt.xlabel(r'$u$ (wavelengths)', fontsize=12)
plt.ylabel(r'$v$ (wavelengths)', fontsize=12)


plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tick_params(axis='both', which='major', labelsize=10)


plt.legend(frameon=True, fontsize=10)
plt.tight_layout()
plt.savefig("uv_coverage.jpg", dpi=300, bbox_inches="tight")  # PNG
plt.show()

```

# UV Plot Analysis

The UV plot above represents a full 24-hour Earth rotation relative to the Sun.  
However, to simulate the complex conjugate property of the **Van Cittert–Zernike Theorem** and the **Hermitian symmetry property**, only a 12-hour rotation was computed.  
The second half of the plot was generated as an exact mirror image using the Hermitian symmetry property.  

It is important to note that this plot corresponds to a **single geometric baseline**, which is the case for the current system configuration.

From the UV coverage, the following can be determined:

1. **Baseline Length**  
   The baseline length $B$ is given by:
   $$
   B = u_{\text{max}} \, \lambda
   $$
   where $\lambda$ is the wavelength of the neutral hydrogen line:
   $$
   \lambda = 21 \, \text{cm} = 0.21 \, \text{m}
   $$
   Thus:
   $$
   B = 93 \times 0.21 = 19.5 \ \text{m} \ \approx \ 20 \ \text{m}
   $$
   This value matches the actual geometric baseline of the system.

2. **Parabolic UV Track**  
   The parabolic trajectory confirms that the baseline is fixed on Earth while the source moves due to Earth's rotation.

3. **Angular Resolution**  
   The angular resolution $\theta_{\text{res}}$ is estimated as:
   $$
   \theta_{\text{res}} \approx \frac{\lambda}{B}
   $$
   Substituting the values:
   $$
   \theta_{\text{res}} = \frac{0.21}{20} = 0.0105 \ \text{rad}
   $$
   Converting to degrees:
   $$
   \theta_{\text{res}} \approx 0.6^\circ
   $$




# Visibility Spectrum
